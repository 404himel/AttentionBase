```python
import cv2
import mediapipe as mp

# Initialize mediapipe face mesh model
mp_face_mesh = mp.solutions.face_mesh
mp_drawing = mp.solutions.drawing_utils

# Initialize webcam
cap = cv2.VideoCapture(0)

# Initialize face mesh model
with mp_face_mesh.FaceMesh(min_detection_confidence=0.5, min_tracking_confidence=0.5) as face_mesh:

    while True:
        # Read frame from webcam
        ret, frame = cap.read()

        if not ret:
            break

        # Convert the frame to RGB
        rgb_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)

        # Process the frame and get facial landmarks
        results = face_mesh.process(rgb_frame)

        if results.multi_face_landmarks:
            for landmarks in results.multi_face_landmarks:
                # Draw the facial landmarks in green color
                for i, landmark in enumerate(landmarks.landmark):
                    # Get the coordinates of each landmark
                    x = int(landmark.x * frame.shape[1])
                    y = int(landmark.y * frame.shape[0])

                    # Draw a green circle at each landmark point
                    cv2.circle(frame, (x, y), 1, (0, 255, 0), -1)

                # Get the position of key facial landmarks: nose (index 1), left eye (index 33), right eye (index 362)
                nose = landmarks.landmark[1]
                left_eye = landmarks.landmark[33]
                right_eye = landmarks.landmark[362]

                # Determine head movement based on the position of the eyes and nose
                head_direction = ""

                # Calculate horizontal position of the eyes and nose (for left-right movement)
                if left_eye.x < nose.x and right_eye.x < nose.x:
                    head_direction = "Look Left"
                elif left_eye.x > nose.x and right_eye.x > nose.x:
                    head_direction = "Look Right"
                # Calculate vertical position of the nose (for up-down movement)
                elif nose.y < left_eye.y and nose.y < right_eye.y:
                    head_direction = "Look Up"
                elif nose.y > left_eye.y and nose.y > right_eye.y:
                    head_direction = "Look Down"
                else:
                    head_direction = "Looking Forward"

                # Display the direction text on the screen
                cv2.putText(frame, head_direction, (50, 50), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)

        # Display the frame with green landmarks and directional text
        cv2.imshow("Facial Landmarks with Direction", frame)

        # Exit if 'q' is pressed
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break

# Release the webcam and close the window
cap.release()
cv2.destroyAllWindows()
```
